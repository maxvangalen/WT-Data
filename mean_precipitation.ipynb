```python
import pandas as pd
```

```python
df = pd.read_csv('DATA/WeatherEvents_Jan2016-Dec2021.csv')
```

```python
df.head()
```

```python
rain = df[['State', 'Precipitation(in)']].groupby('State').sum()
stations = df[['State', 'AirportCode']].groupby('State').nunique()
```

```python
full = rain['Precipitation(in)'] / stations['AirportCode']
full = full.reset_index()
full.columns = ['State', 'Precipitation']
```

```python
full
```

```python
full.to_csv('DATA/Precipitation_Per_State.csv', index=False)
```
